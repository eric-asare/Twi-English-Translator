```python
#load model and tokenizer
from transformers import MarianMTModel, MarianTokenizer

model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-tw")
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-tw")
```

```python
#run this to enter your english text eg: Thank you --> Me da ase

englishText = input("Your sentence: ")
```

```python
# your entered text is stored here
text = englishText

# Encode to input token ids
inputs = tokenizer.encode(text, return_tensors="pt") 
# Generate output token ids
outputs = model.generate(inputs) 
# Decode output token ids to actual output tokens
decoded_output = [tokenizer.convert_ids_to_tokens(int(outputs[0][i])) for i in range(len(outputs[0]))] 

#Print original text
print("English:", text)

#trim translated output

size = len(decoded_output) 
trimEnd = size - 1
finalOutput = ' '.join(decoded_output[1:trimEnd])

#remove trailing _
toRemove = finalOutput[0]
finalOutput = finalOutput.replace(toRemove, "")

print("Twi: ", finalOutput)
```
